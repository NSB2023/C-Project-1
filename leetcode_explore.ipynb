```python
#217
'''
Question: Contains Duplicate
Given an integer array nums, return true if any value appears at least twice in
the array, and return false if every element is distinct.
'''

class Solution(object):
    def containsDuplicate(self, nums):
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
s = Solution()
print(s.containsDuplicate([1, 2, 2, 1]))
```

```python
#242
'''
valid anagram problem solv:
Given two strings s and t, return true if t is an anagram of s, and false otherwise.
'''
class solution:
  def isAnagram(self,s,t):
    d1={}
    d2={}

    for i in s:
      d1[i]=d1.get(i,0) +1
    for j in t:
      d2[j]=d2.get(j,0) +1

    return d1==d2

solution = solution()
print(solution.isAnagram('cat','rat'))
print(solution.isAnagram('cat','act'))

```

```python
# 1
'''
Two Sum
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
'''
class Solution(object):
    def twoSum(self, nums, target):
      for i in range(0,len(nums),1):
        for j in range(i+1,len(nums),1):
          if nums[i]+nums[j] == target:
            return [i,j]
obj=Solution()
print(obj.twoSum([2,7,11,15],9))
```

```python
#49  - Group Anagram !
from collections import defaultdict
class solution:
  def groupanagram(self,strs):
    d=defaultdict(list)
    for i in strs:
      sorted_str=sorted(i)
      final_sorted_str ="".join(sorted_str)
      d[final_sorted_str].append(i)
    return list(d.values())
obj=solution()
print(obj.groupanagram(["act","pots","tops","cat","stop","hat","dhfdhfgd"]))
```

```python
#128. Longest Consecutive Sequence
# Hashset
class Solution(object):
    def longestConsecutive(self, nums):
        count=0
        s=set(nums)
        for i in s:
          if (i - 1) not in s:
             length = 1
             while i+length in s:
                length +=1
             count = max(length,count)
        print('longest consecutive sequence :',count)

obj=Solution()
obj.longestConsecutive([100, 4, 200, 1, 3, 2,2])
```
